## Analise exploratoria Nutricional

```python
# @title
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    r"/content/drive/MyDrive/PNS2019/pns2019.csv",
    sep=",",
    low_memory=False
)

# Limpar nomes de colunas (evita erro escondido)
df.columns = df.columns.str.strip()

# ##########################################
#-----> AQUI EMBAIXO COLOCA O ATRIBUTO DA SUA DOENCA
##########################################
df = df.rename(columns={"Q092": "depressao"})


# Tirei o filtro de idade
# Pessoas com depressão
depressao = df.query("depressao == 1") # and C008 < 30 and C008 > 17")

# Pessoas sem nenhuma doença
sem_doenca = df.query(
    "Q00201 == 2 and Q03001 == 2 and Q060 == 2 and Q06306 == 2 and "
    "Q068 == 2 and Q074 == 2 and Q079 == 2 and Q088 == 2 and "
    "Q11604 == 2 and Q120 == 2 and Q124 == 2 and Q128 == 2 and "
    "depressao == 2 and Q084 == 2 and Q11006 == 2"
)

# Base final de análise, filtrada
df_analise = pd.concat([depressao, sem_doenca])






print("Categorias (dias por semana) Baixo: \n 0 a 2 dias \n Médio: 3 a 5 dias \n Alto: 6 a 7 dias")

# =========================
# 1. Lista de atributos de alimentos por semana
# =========================
atributos = [
    'P006', 'P00901', 'P01101', 'P013', 'P015',
    'P01601', 'P018', 'P02001', 'P02002',
    'P023', 'P02501', 'P02602', 'P02801'
]

# =========================
# 1.1 Nome legível dos atributos
# =========================
nomes = {
    'P006': 'Feijão',
    'P00901': 'Verduras/Legumes',
    'P01101': 'Carne Vermelha',
    'P013': 'Frango',
    'P015': 'Peixe',
    'P01601': 'Suco Natural',
    'P018': 'Frutas',
    'P02001': 'Suco Industrializado',
    'P02002': 'Refrigerante',
    'P023': 'Leite',
    'P02501': 'Doces',
    'P02602': 'Substituição de Refeição',
    'P02801': 'Álcool'
}

# =========================
# 2. Criar variável de doença
# =========================
df_analise['doenca'] = df_analise['depressao'].apply(lambda x: 1 if x == 1 else 0) #se for 1, tem a doença, de resto considera q nn tem

# =========================
# 3. Loop
# =========================
resultados = {}

for atributo in atributos:

    df_temp = df_analise[[atributo, 'doenca']].copy()

    # Remover valores inválidos
    df_temp = df_temp[df_temp[atributo] != 9]

    # Criar categorias
    df_temp[atributo + '_cat'] = pd.cut(
        df_temp[atributo],
        bins=[-1, 2, 5, 7],
        labels=['Baixo', 'Médio', 'Alto']
    )

    # Tabela, aqui calcula a P(doenc​a∣categoria)
    tabela = pd.crosstab(
        df_temp[atributo + '_cat'],
        df_temp['doenca'],
        normalize='index'
    )

    resultados[atributo] = tabela

    if 1 in tabela.columns:
        prob_doenca = tabela[1]

        nome_legivel = nomes.get(atributo, atributo)

        # Gráfico
        plt.figure()
        prob_doenca.plot(kind='bar')
        plt.title(f'Probabilidade de doença vs {nome_legivel}')
        plt.xlabel('Consumo')
        plt.ylabel('Probabilidade de doença')
        plt.xticks(rotation=0)
        plt.show()

        # Print bonito
        print(f"\n=== {nome_legivel} ({atributo}) ===")
        print(prob_doenca)

# import matplotlib.pyplot as plt

prob_doenca.plot(kind='bar')
plt.title('Probabilidade de doença por consumo de refrigerante')
plt.xlabel('Consumo de refri')
plt.ylabel('Probabilidade de doença')
plt.show()

#-------- AQUI FAZ O RANKING

ranking = []

for atributo, tabela in resultados.items():
    if 1 in tabela.columns:

        # Garantir que todas categorias existem
        if all(cat in tabela.index for cat in ['Baixo', 'Médio', 'Alto']):

            baixo = tabela.loc['Baixo', 1]
            alto = tabela.loc['Alto', 1]

            impacto = alto - baixo

            nome_legivel = nomes.get(atributo, atributo)

            ranking.append((nome_legivel, atributo, impacto))

# Ordenar do maior impacto pro menor
ranking = sorted(ranking, key=lambda x: x[2], reverse=True)

# Print bonito
print("\n=== Ranking de impacto (Alto - Baixo) ===\n")

for nome, cod, impacto in ranking:
    print(f"{nome} ({cod}) → {impacto:.4f}")
```
